```python
lines = open('input.txt').read().splitlines()
```

```python
dirs = lines.pop(0)
lines.pop(0)
dirs
```

```python
m = {}
for line in lines:
    node = line.split()[0]
    left = line.split()[2][1:-1]
    right = line.split()[3][:-1]
    m[node] = (left,right)
```

```python
m
```

```python
from itertools import cycle

c = 'AAA'
count = 0

pool = cycle(dirs)
for d in pool:
    if c == 'ZZZ':
        break
        
    count += 1
    if d == 'L':
        c = m[c][0]
    elif d == 'R':
        c = m[c][1]
    else:
        print('ERROR')
        break
        
print(count)
```

```python
cs = [x for x in m.keys() if x[-1] == 'A']

def end_criteria(cs):
    n = len(cs)
    return (n == len([x for x in cs if x[-1] == 'Z']))

count = 0
pool = cycle(dirs)
for d in pool:
    if end_criteria(cs):
        break
    
    count += 1
    
    if d == 'L':
        for i,c in enumerate(cs):
            cs[i] = m[c][0]            
    elif d == 'R':
        for i,c in enumerate(cs):
            cs[i] = m[c][1]
```
