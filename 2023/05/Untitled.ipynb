```python
filename = 'test.txt'

seeds, *blocks = open(filename).read().split("\n\n")
seeds = list(map(int, seeds.split(":")[1].split()))
```

```python
bb = []
for k,block in enumerate(blocks):
   ranges = []
   for line in block.splitlines()[1:]:
     ranges.append( list(map(int, line.split())) )
   bb.append(ranges)
```

```python
seed_ranges = list(zip(seeds[::2], seeds[1::2]))
```

```python
seed_ranges
```

```python
for ranges in bb:
    new = []
    while len(seed_ranges) > 0:
        s0,s1 = seed_ranges.pop()
        S_min = s0
        S_max = s0+s1-1

        for r0, r1, r2 in ranges:
            R_min = r1
            R_max = r1+r2-1

            # Case 1 R to left of S or R to right of S
            if (R_max < S_min) or (S_max < R_min):
                pass

            # Case 2: intersect on left
            elif (R_min < S_min) and (R_min < S_min <= R_max) and (R_max <= S_max):
                # Two outputs
                seed_ranges.append([R_min, S_min-1]) # This goes on the list to process
                new.append([S_min - r1 + r0, R_max - r1 + r0])   # this gets mapped now
                break

            # Case 4: intersect on right
            elif (S_min <= R_min <= S_max) and (S_max > R_max):
                # Two outputs
                new.append([R_min - r1 + r0, S_max - r1 + r0])   # this gets mapped now
                seed_ranges.append([S_max+1, R_max]) # This goes on the list to process
                break

            # Case 3: S inside R
            elif S_min <= R_min and R_max <= S_max:
                # One output
                new.append([S_min - r1 + r0, S_max - r1 + r0])   # this gets mapped now
                break

            # Case 5: S inside of R
            elif (R_min <= S_min <= R_max) and (R_min <= S_max <= R_max):
                # Three outputs
                seed_ranges.append([R_min, S_min-1]) # This goes on the list to process   # Left
                new.append([S_min - r1 + r0, S_max - r1 + r0])   # this gets mapped now               # Center
                seed_ranges.append([S_max+1, R_max]) # This goes on the list to process   # Right
                break

        else:
            # source didn't overlap with any of the ranges
            new.append([S_min, S_max])

    seed_ranges = new
    
        
#    print(S_min, S_max)
```
